# Cálculo variacional

## Funcionais

$$I=\int_{x_1}^{x_2}F(y(x),x)dx$$

O Funcional $I$ não depende de $x$, pois estamos integrando em todo $x$ e, após a integração
o $x$ desaparece. Porém, $I$ depende do tipo de função $y$. $I$ tem um valor para cada
função.

Exemplo, considerando $y(x)=x^2$ e $x_1=1$ e $x_2=2$. Para $F(y(x),x)=xy(x)$,

$$I=\int_{1}^{2}xx^2dx=\frac{x^4}{4}\Big|_1^2=\frac{2^4}{4}-\frac{1}{4}=\frac{15}{4}$$

Se, $y(x)=x^3$,

$$I=\int_{1}^{2}xx^3dx=\frac{x^5}{5}\Big|_1^2=\frac{2^5}{5}-\frac{1}{5}=\frac{31}{5}$$

$I$ é uma função de outra função. $I$ é chamado de _funcional_ de $y$.

## Extremos da função y(x)

Para qual $x$ temos $dy=0$ para $y(x)$?

$$dy=y(x+dx)-y(x)$$

Expandindo $y(x+dx)$ em série de Taylor:

$$dy=y(x)+\frac{dy}{dx}dx+O(x^2)-y(x)$$

$$dy=\frac{dy}{dx}dx$$

Como $dx\neq 0$, e na condição de extremo $dy=0$, temos que,

$$\frac{dy}{dx}=0$$

Patindo da definição de $I$:

$$I = \int_{x_1}^{x_2}F(y, x)dx$$

Assim,

$$\delta I = \int_{x_1}^{x_2}F(y+\delta y, x)dx- \int_{x_1}^{x_2}F(y, x)dx$$

Expandindo $F(y+\delta y, x)$ em série de Taylor:

$$F(y+\delta y,x)=F(y,x)+\frac{\partial F(y,x)}{\partial y}\delta y$$

Então,

$$\delta I=\int_{x_1}^{x_2}\left[ F(x,y)+\frac{\partial F(y,x)}{\partial y}\delta y\right]dx
-\int_{x_1}^{x_2} F(x,y)dx$$

$$\delta I=\int_{x_1}^{x_2}\frac{\partial F(y,x)}{\partial y}\delta ydx$$

A condição de extremo implica que $\delta I=0$, portanto,

$$\int_{x_1}^{x_2}\frac{\partial F(y,x)}{\partial y}\delta ydx=0 \to \frac{\partial F}{\partial y}=0$$

## Equação de Euler-Lagrange

Considere agora o caso em que:

$$F=F(y(x),y'(x),x)$$

$$I=I[y,y']$$

Primeiro, temos que,

$$\delta I=\int_{x_1}^{x_2}\left(\frac{\partial F}{\partial y}\delta y+\frac{\partial F}{\partial y'}\delta y'\right)dx$$

Pois,

$$\delta I = \int_{x_1}^{x_2}F(y+\delta y, y'+\delta y',x)dx - \int_{x_1}^{x_2}F(y,y',x)dx$$

Expandindo em série de Taylor 2D:

$$F(y+\delta y, y'+\delta y',x)\approx F(y,y',x)+\frac{\partial F}{\partial y}\delta y+\frac{\partial F}{\partial y'}\delta y'+ \dots$$

Assim,

$$\delta I=\int_{x_1}^{x_2}\left[ F(y,y',x)+\frac{\partial F}{\partial y}\delta y+\frac{\partial F}{\partial y'}\delta y'\right]dx - \int_{x_1}^{x_2}F(y,y',x)dx$$

$$\delta I=\int_{x_1}^{x_2}\left[ \frac{\partial F}{\partial y}\delta y+\frac{\partial F}{\partial y'}\delta y'\right]dx $$

Desenvolvendo o segundo termo da integral:

$$\int_{x_1}^{x_2}\frac{\partial F}{\partial y'}\delta y'dx=
\int_{x_1}^{x_2}\frac{\partial F}{\partial y'}\delta \frac{dy}{dx}dx=\int_{x_1}^{x_2}\frac{\partial F}{\partial y'}\frac{d}{dx}\delta ydx=\frac{\partial F}{\partial y'}\delta y\Big|_{x_1}^{x_2}-\int_{x_1}^{x_2}\delta y\frac{d}{dx}\frac{\partial F}{\partial y'}dx$$

Pois, pela regra da integração por partes:

$$\int udv= uv - \int vdu$$

$$u=\frac{\partial F}{\partial y'}$$

$$dv=\frac{d}{dx}\delta y dx$$

$$du=\frac{d}{dx}\frac{\partial F}{\partial y'}dx$$

$$v=\delta y$$

Então,

$$\int_{x_1}^{x_2}\frac{\partial F}{\partial y'}\delta y'dx=\frac{\partial F}{\partial y'}\delta y\Big|_{x_1}^{x_2}-\int_{x_1}^{x_2}\delta y\frac{d}{dx}\frac{\partial F}{\partial y'}dx$$

Como os pontos $p_1$ e $p_2$ são fixos, $\delta y|_{x_1}^{x_2}=0$,

$$\delta I=\int_{x_1}^{x_2}\left(\frac{\partial F}{\partial y}-\frac{d}{dx}\frac{\partial F}{\partial y'}\right)\delta ydx$$

Considerando a condição de extremo $\delta I=0$:

$$\frac{\partial F}{\partial y}-\frac{d}{dx}\frac{\partial F}{\partial y'}$$

Esta equação foi obtida por Euler em 1744. Ela é conhecida como _Equação de Euler-Lagrange_
devido ser ela a base da formulação Lagrangiana da mecânica clássica. Esta representa a
condição de extremo para o funcional $I$, onde $F=F(y,y')$.

## A formulação Lagrangiana

$$L=L(q_1,q_2,\dots,q_N;\dot{q_1},\dot{q_2},\dots,\dot{q_N},t)$$

Compactamente,

$$L=L(q,\dot{q},t)$$

Representa a Lagrangiana do sistema e a quantidade $S$ a seguir
é chamada de ação:

$$S=\int_{t_1}^{t_2}L(q,\dot{q},t)dt$$

_O princípio de Hamilton, também denominado de mínima ação,
estabelece o seguinte: A evolução do sistema da configuração 1
para a configuração 2 é tal que a ação é um mínimo._

Pelos nossos estudos de cálculo variacional, temos que para uma
partícula, e N coordenadas generalizadas,

$$\delta S=\int_{t_1}^{t_2}\sum_{i=1}^{N}\left(\frac{\partial L}{\partial q_i}\delta q_i+\frac{\partial L}{\partial\dot{q_i}}\delta \dot{q_i}\right)dt$$

Assim,

$$\delta S=\int_{t_1}^{t_2}\sum_{i=1}^{N}\left(\frac{\partial L}{\partial q_i}-\frac{d}{dt}\frac{\partial L}{\partial\dot{q_i}} \right)\delta q_i(t)dt$$

Como os $\delta q_i(t)$ são funções arbitrárias de $t$,
a condição de mínimo, $\delta S=0$, é satisfeita se

$$\sum_{i=1}^{N}\left(\frac{\partial L}{\partial q_i}-\frac{d}{dt}\frac{\partial L}{\partial\dot{q_i}} \right)\delta q_i(t)=0$$

Assim, se não houver relação de vínculo entre os $q_i$'s, temos
que os diversos $\delta q_i$'s são independentes. Neste caso,
e somente neste caso, temos

$$\frac{\partial L}{\partial q_i}-\frac{d}{dt}\frac{\partial L}{\partial\dot{q_i}}=0$$

$$i=1,2,\dots,N$$

Esta é a equação de _Euler-Lagrange_. Para usar esta expressão, precisamos conhecer a
lagrangiana $L$ do sistema, uma função escalar que caracteriza o sistema estudado.

## A formulação Hamiltoniana

A função Hamiltoniana,

$$H=\sum_i p_i\dot{q_i}-L(q,\dot{q},t)$$

Na formulação Lagrangiana, as variáveis independentes são $(q,\dot{q},t)$. No caso da
formulação Hamiltoniana o conjunto de variáveis independentes é dado por $(q,p,t)$. Onde $p_i=\partial L/\partial \dot{q_i}$ e $\dot{p_i}=\partial L/\partial q_i$.

$$dL=\sum_i \frac{\partial L}{\partial q_i}dq_i+\sum_i \frac{\partial L}{\partial\dot{q_i}}d\dot{q_i}+\frac{\partial L}{\partial t}dt$$

$$dL=\sum_i \frac{\partial L}{\partial q_i}dq_i+\sum_i p_id\dot{q_i}+\frac{\partial L}{\partial t}dt$$

Lembrando:

$$d(p_i\dot{q_i})=\dot{q_i}dp_i+p_id\dot{q_i}$$

Então,

$$dH=d\left(\sum_ip_i\dot{q_i}-L\right)=d\left(\sum_ip_i\dot{q_i}\right)-dL$$

E,

$$d\left(\sum_ip_i\dot{q_i}\right)=\sum_i\dot{q_i}dp_i+\sum_ip_id\dot{q_i}$$

Logo,

$$d\left(\sum_ip_i\dot{q_i}-L\right)=d\left(\sum_ip_i\dot{q_i}\right)-dL$$

Assim,

$$d\left(\sum_ip_i\dot{q_i}-L\right)=
\sum_i\dot{q_i}dp_i+\sum_ip_id\dot{q_i}-\sum_i\frac{\partial L}{\partial q_i}dq_i-\sum_ip_id\dot{q_i}-\frac{\partial L}{\partial t}dt$$

Por fim,

$$d\left(\sum_ip_i\dot{q_i}-L\right)=dH=
\sum_i\dot{q_i}dp_i-\sum_i\frac{\partial L}{\partial q_i}dq_i-\frac{\partial L}{\partial t}dt$$

Oque mostra nitidamente que $H=H(q,p,t)$. Observe que para se chegar a essa conclusão,
não foi necessário usar a _Equação de Euler-Lagrange_, apenas usamos a definição de
_momento canônico_, $p_i$. Isto significa que $H=H(q,p,t)$ independentemente do _princípio
de Hamilton_, ou seja, mesmo fora da trajetória clássica.

Substituindo $\frac{\partial L}{\partial q_i}$ por $\dot{p_i}$, temos,

$$dH=
\sum_i\dot{q_i}dp_i-\sum_i\dot{p_i}dq_i-\frac{\partial L}{\partial t}dt$$

Se,

$$dH=
\sum_i\frac{\partial H}{\partial q_i}dq_i+\sum_i\frac{\partial H}{\partial p_i}dp_i+\frac{\partial H}{\partial t}dt$$

Assim,

$$\dot{q_i}=\frac{\partial H}{\partial p_i}$$

$$\dot{p_i}=-\frac{\partial H}{\partial q_i}$$

Estas são as chamadas _Equações de Hamilton_.

Ambos os formalismos, Lagrangiano e Hamiltoniano se baseiam no _princípio de Hamilton_: No
formalismo Lagrangiano temos $N$ equações diferenciais de segunda ordem, desenvolvidas no
espaço de $N$ coordenadas generalizadas, chamado de _espaço das configurações_. No
Hamiltoniano, são $2N$ equações de primeira ordem e o espaço possui $2N$ coordenadas
($N$ coordenadas generalizadas e $N$ momentos) chamado de _espaço das fases_.

Partindo do princípio de Hamilton.

$$\delta\int_{t_1}^{t_2}L(q,\dot{q},t)dt=0$$

Isto é,

$$\delta\int_{t_1}^{t_2}\left[\sum_ip_i\dot{q_i}-H(q,p,t)\right]dt$$

Desenvolvendo a relação acima,

$$\int_{t_1}^{t_2}\sum_i\left[p_i\delta\dot{q_i}+\dot{q_i}\delta p_i-\frac{\partial H}{\partial q_i}\delta q_i-\frac{\partial H}{\partial p_i}\delta p_i\right]dt=0$$

Como nos extremos, $t_1$ e $t_2$, $\delta q_i=0$, temos

$$\int_{t_1}^{t_2}p_i\delta\dot{q_i}dt=-\int_{t_1}^{t_2}\dot{p_i}\delta q_idt$$

Pois, pela integração por partes,

$$\int udv=uv-\int vdu$$

Então,

$$\int_{t_1}^{t_2}p_i\delta \dot{q_i}dt=p_i\delta q_i\Big|_{t_1}^{t_2}-\int_{t_1}^{t_2}\dot{p_i}\delta q_idt$$

Onde:

$$u=p_i$$

$$dv=\delta\dot{q_i}dt$$

$$du=\dot{p_i}dt$$

$$v=\delta q_i$$

Substituindo na integral anterior:

$$\int_{t_1}^{t_2}\sum_i\left[p_i\delta\dot{q_i}+\dot{q_i}\delta p_i-\frac{\partial H}{\partial q_i}\delta q_i-\frac{\partial H}{\partial p_i}\delta p_i\right]dt=0$$

Desta forma:

$$\int_{t_1}^{t_2}\sum_i\left[-\dot{p_i}\delta q_i+\dot{q_i}\delta p_i-\frac{\partial H}{\partial q_i}\delta q_i-\frac{\partial H}{\partial p_i}\delta p_i\right]dt=0$$

Arrumando e isolando $\delta p_i$ e $\delta q_i$ em evidência,

$$\int_{t_1}^{t_2}\sum_i\left[\left(\dot{q_i}-\frac{\partial H}{\partial p_i}\right)\delta p_i-\left(\dot{p_i}+\frac{\partial H}{\partial q_i}\right)\delta q_i\right]dt=0$$

O Hamiltoniano,

$$H=\sum_i p_i\dot{q_i}-L$$

Fazendo a derivada em relação à $p_k$:

$$\frac{\partial H}{\partial p_k}=\dot{q_k}+\sum_ip_i\frac{\partial\dot{q_i}}{\partial p_k}-\sum_i\frac{\partial L}{\partial\dot{q_i}}\frac{\partial\dot{q_i}}{\partial p_k}=\dot{q_k}$$

$$\dot{q_i}=\frac{\partial H}{\partial p_i}$$

Esta é a primeira equação de Hamilton. Retomando o resultado anterior,

$$\int_{t_1}^{t_2}\sum_i\left[\left(\dot{q_i}-\frac{\partial H}{\partial p_i}\right)-\left(\dot{p_i}+\frac{\partial H}{\partial q_i}\right)\delta q_i\right]dt=0$$

$$\int_{t_1}^{t_2}\sum_i\left(\dot{p_i}+\frac{\partial H}{\partial q_i}\right)\delta q_idt=0$$

Daí surge que,

$$\dot{p_i}=-\frac{\partial H}{\partial q_i}$$

Que é a segunda equação de Hamilton.
