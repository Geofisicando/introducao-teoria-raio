### Derivao das equaes x(p) e t(p) na forma discreta

Considerando que a velocidade varia apenas com a profundidade,
a vagarosidade horizontal  a mesma para todas as camadas:

$$p=\frac{\sin\theta_1}{v_1}=\frac{\sin\theta_2}{v_2}=
\frac{\sin\theta_3}{v_3}=\dots=\frac{\sin\theta_n}{v_n}$$

Assim, para uma camada i,

$$\sin\theta_i=pv_i$$

A distncia horizontal $\Delta x_i$ ser:

$$\Delta x_i=\Delta z_i\tan\theta_i$$

Pois,

$$\tan\theta_i=\frac{\Delta x_i}{\Delta z_i}$$

Assim,

$$X=\Delta z_1\tan\theta_1+\Delta z_2\tan\theta_2+\Delta z_3\tan\theta_3+\dots+\Delta z_n\tan\theta_n$$


Como o raio  simtrico em relao ao _turning point_:

$$X=2\sum_{i=1}^{n}\Delta z_i\tan\theta_i$$

Da identidade trigonomtrica:

$$\sin^2\theta_i+\cos^2\theta_i=1$$

$$\cos^2\theta_i=1-\sin^2\theta_i$$

$$\cos\theta_i=\sqrt{1-\sin^2\theta_i}$$

E da definio de $p$:

$$p=\frac{\sin\theta_i}{v_i}$$

$$\sin\theta_i=pv_i$$

Ento,

$$\tan\theta_i=\frac{\sin\theta_i}{\cos\theta_i}=\frac{pv_i}{\sqrt{1-p^2v^2_i}}$$

Substituindo na definio de $X$:

$$X=2\sum_{i=1}^{n}\Delta z_i\tan\theta_i=
2\sum_{i=1}^{n}\Delta z_i\frac{pv_i}{\sqrt{1-p^2v^2_i}}$$

Definindo em termos de $u_i$:

$$u_i=1/v_i$$

Logo,

$$X=2\sum_{i=1}^{n}\Delta z_i\frac{pv_i}{\sqrt{1-p^2v^2_i}}$$

$$X=2\sum_{i=1}^{n}\Delta z_i\frac{p}{u_i\sqrt{1-\frac{p^2}{u^2_i}}}$$

$$X=2\sum_{i=1}^{n}\Delta z_i\frac{p}{u_i\sqrt{\frac{u^2_i}{u^2_i}-\frac{p^2}{u^2_i}}}$$

Por fim:

$$X=2p\sum_{i=1}^{n}\frac{\Delta z_i}{\sqrt{u^2_i-p^2}}$$

Considerando a definio de $v_i$:

$$v_i=\frac{\Delta s_i}{\Delta T_i}$$

Ento,

$$\Delta T_i=\frac{\Delta s_i}{v_i}$$

Lembrando que:

$$\cos\theta_i=\frac{\Delta z_i}{\Delta s_i}$$

Assim,

$$\Delta s_i=\frac{\Delta z_i}{\cos\theta_i}$$

Logo, podemos definir $\Delta T_i$:

$$\Delta T_i=\frac{\Delta z_i}{v_i\cos\theta_i}$$

Para todas as $i$ camadas, o tempo total $T$ ser:

$$T=\frac{\Delta z_1}{v_1\cos\theta_1}+\frac{\Delta z_2}{v_2\cos\theta_2}+
\frac{\Delta z_3}{v_3\cos\theta_3}+\dots+\frac{\Delta z_n}{v_n\cos\theta_n}$$

Como o raio  simtrico em relao ao _turning point_:

$$T=2\sum_{i=1}^{n}\frac{\Delta z_i}{v_i\cos\theta_i}$$

Utilizando a identidade trigonomtrica para o $\cos\theta_i$:

$$T=2\sum_{i=1}^{n}\frac{\Delta z_i}{v_i\sqrt{1-\sin^2\theta_i}}=
2\sum_{i=1}^{n}\frac{\Delta z_i}{v_i\sqrt{1-p^2v^2_i}}$$

Definindo em termos de $u_i$:

$$u_i=1/v_i$$

Logo,

$$T=2\sum_{i=1}^{n}\frac{\Delta z_iu_i}{\sqrt{1-\frac{p^2}{u^2_i}}}$$

$$T=2\sum_{i=1}^{n}\frac{\Delta z_iu_i}{\sqrt{\frac{u^2_i}{u^2_i}-\frac{p^2}{u^2_i}}}
=2\sum_{i=1}^{n}\frac{\Delta z_iu_i}{\frac{1}{u_i}\sqrt{u^2_i-p^2}}$$

Multiplicando o numerador e o denominador por $u_i$:

$$T=2\sum_{i=1}^{n}\frac{\Delta z_iu_i}{\frac{1}{u_i}\sqrt{u^2_i-p^2}}\frac{u_i}{u_i}$$

Ento,

$$T=2\sum_{i=1}^{n}\frac{\Delta z_iu^2_i}{\sqrt{u^2_i-p^2}}$$
