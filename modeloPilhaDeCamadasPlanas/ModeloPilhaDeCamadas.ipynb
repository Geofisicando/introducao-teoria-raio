\begin{equation}
 \label{eq:1.15}
 x(p)=2p\int^{z_p}_{0} \frac{dz}{(u^2(z)-p^2)^{1/2}}
\end{equation}

Expressando a integral acima na forma discreta:

\begin{equation}
 \label{eq:1.16}
 x(p)=2p\sum_i \frac{\Delta z_i}{(u^2_i-p^2)^{1/2}}
\end{equation}

Vlido sob a condio:

\begin{equation}
\label{eq:1.17}
u_i > p
\end{equation}



\begin{equation}
 \label{eq:1.20}
 t(z_1,z_2,p)=\int^{z_1}_{z_2} \frac{u^2(z)}{(u^2(z)-p^2)^{1/2}} dz
\end{equation}

Como o raio  simtrico em relao ao \textit{turning point}, a distncia total percorrida em $x$ ser:

\begin{equation}
 \label{eq:1.21}
 t(p)=2\int^{z_p}_{0} \frac{u(z)^2}{(u^2(z)-p^2)^{1/2}} dz
\end{equation}

Expressando a integral acima na forma discreta:

\begin{equation}
 \label{eq:1.22}
 t(p)=2\sum_i \frac{u^2_i\Delta z_i}{(u^2_i-p^2)^{1/2}}
\end{equation}

Vlido tambm sob a condio da Equao \ref{eq:1.17}. Esta condio existe para garantir que o produto $(u^2_i-p^2)^{1/2}$
no se torne um nmero imaginrio.
